### Visualizing irrational numbers

```python
from mpmath import mp
import numpy as np
import matplotlib.pyplot as plt
import re
from matplotlib import cm
%matplotlib inline

def plot_irrational(x, linewidth=0.015, alpha =0.6, wrap = 0.5, ax = None, linecolor = 'b'):
    """
    Parameters
    ----------
    
    x          :     float, number to be plotted
    
    linewidth  :     float, specifies linewidth of final plot
    
    alpha      :     float, specifies alpha of final plot
    
    wrap       :     float, specifies over what percentage of 2*pi radians to plot the "nodes"
                     which results in each 0-9 digit being represented by an arc of length 2*wrap*pi / 10
    
    ax         :     axes object, if given will plot on the specified axis
    
    linecolor  :     specifies linecolor of the plot
    
    
    Results
    -------
    
    Plot of x either shown immediately or assigned to the specified axes.
    """
    
    x_ = str(int(x)) + str(x - int(x))[2:]
    n = len(x_)
    mp.dps = 8
    r_part = np.pi / 10.0
    
    node_list = np.zeros(n)
    node_list[0] = r_part*int(x_[0])
    
    for i in range(1,n):
        """
        Based on how the default matplotlib function works for plotting
        the code can be thought of as creating a graph where edges
        are drawn between nodes listed next to eachother in a list
    
        G.add_node(new_node)
        G.add_edge(old_node, new_node)
    
        where old_node = node_list[i-1] and new_node = node_list[i]
        """
        node_list[i] = 2.0*r_part*int(x_[i]) + 2.0* wrap * r_part*i/n
        
    x = np.cos(node_list)
    y = np.sin(node_list)
    
    if ax is None:
        fig = plt.figure(figsize = (10,10))
        ax = fig.add_subplot(111)
        
    ax.plot(x ,y, 
            linewidth=linewidth, 
            alpha = alpha, 
            color = linecolor)
        
    ax.set_xlim(-1.0, 1.0)
    ax.set_ylim(-1.0, 1.0)
    ax.set_axis_off()
```

```python
mp.dps = 10000
pi = mp.pi

plot_irrational(pi, linecolor = 'red')
```

```python
mp.dps = 10000
sqrt_2 = mp.sqrt(2.0)

plot_irrational(sqrt_2, wrap = 6.7)
```

```python
mp.dps = 10000
e = mp.exp(1.0)

plot_irrational(e, wrap = 0.25)
```

```python
mp.dps = 10000
e = mp.exp(1.0)

plot_irrational(e, wrap = 0)
```
