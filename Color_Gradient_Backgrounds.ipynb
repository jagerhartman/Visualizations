```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
#import matplotlib.colors.rgb as rgb

size = 5
fig= plt.figure(figsize=(3*size, 3*size))
ax = fig.add_subplot(1,1,1, aspect='equal')
Y = range(3*size*2)
X = range(4*size*2)
l=[]
for x in X:
    for y in Y:
        color_ratio = int((y+1)/(y+x+1)*5)/5
        l.append(color_ratio)
        #print (color_ratio)
        ax.add_patch(
        patches.Rectangle(
            (x, y),   # (x,y)
            1.0,          # width
            1.0,          # height
            
            fc = (color_ratio,0, 1-color_ratio),
            #edgecolor = (1,0,1)
        ))
zoom_ratio = 0.0
ax.set_ylim(zoom_ratio*Y[-1],(1-zoom_ratio)*Y[-1])
ax.set_xlim(zoom_ratio*X[-1],(1-zoom_ratio)*X[-1])
ax.axis('off')
plt.show()
print(np.unique(l))
```

```python
import numpy as np
import matplotlib
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 14}

matplotlib.rc('font', **font)

b = ['Byte', 'KiloByte', 'MegaByte', 'GigaByte', 'TeraByte', 'PetaByte', 'ExaByte']
x = np.linspace(1, 7)
y = 2**(10*x)


fig = plt.figure(figsize=(10,(9/16)*10))
fig.patch.set_facecolor((1.0,1,1.0))
ax = fig.add_subplot(111)
ax.plot(x,y, 'b', linewidth=3)
ax.set_xticks([1,2,3,4,5,6,7])
ax.set_xticklabels(b[:])
ax.set_yticks(range(2**10*5, 5))
ax.tick_params(axis='y', colors='b')
ax.tick_params(axis='x', colors='b')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('b')
ax.spines['bottom'].set_color('b')
ax.spines['left'].set_linewidth(3)
ax.spines['bottom'].set_linewidth(3)

ax.axvline(3.01, 0, 0.8, linestyle = '--',color = 'b')
ax.axvline(4 + (3.5/100), 0, 0.8, linestyle = '--',color = 'b')
ax.axvline(6 + (25/100), 0, 0.8, linestyle = '--',color = 'b')
ax.axvline(5 + (8/100), 0, 0.8, linestyle = '--',color = 'b')
ax.axvline(7 + (8/100), 0, 1.3, linestyle = '--',color = 'b')

ax.set_facecolor((1.0,1.0,1.0))

ax.set_ylabel('Bytes', color = 'b', weight='bold')

plt.show()
```
